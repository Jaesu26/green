# 01. matplotlib 기본 사용
`-` `matplotlib` 라이브러리를 이용해서 그래프를 그려보자

`-` 그 중에서도 `pyplot` 사용할 것 임

## 기본 그래프 

`-` `pyplot`을 이용해서 간단하게 시각화 가능

`-` `plt.plot()` 에 `하나의 숫자 리스트`를 입력함으로써 아래와 같은 그래프가 그려짐

`-` `plt.plot()`은 `하나의 리스트`만 `입력`되면 `리스트의 값들이 y 값`들이라고 가정하고, `x 값 [0, 1, 2, 3] (default)`을 `자동으로 만듦`

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.show()
```

`-` `plot()` 함수는 다양한 기능을 포함하고 있어서, `임의의 개수`의 `인자`를 `받을 수 있음`

`-` 아래와 같이 `두 개의 리스트`를 입력하면 `순서대로 x, y값`으로 생각하고 그래프를 나타냄

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40])
```

`-` `plt.show()` 는 `화면에 표시`하는 기능을 하는데 `Jupyter`를 사용할 때는 `자동으로 표시`되기 때문에 호출할 필요가 없음

## 스타일 지정하기 
`-` `x, y 값 인자`에 대해 `선의 색상`과 `형태`를 지정하는 `포맷 문자열`을 `세번째 인자`에 `입력`할 수 있음 ---> 후에 더 자세히 알아보자

`-` `axis()`를 이용해서 축의 범위 `[xmin, xmax, ymin, ymax]`를 지정함

`-` `linewidth`를 통해 `선 두께` 조절

`-` `markersize`를 통해 `마커 사이즈` 조절

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro--', linewidth = 3, markersize = 9)
plt.axis([0, 6, 0, 20])
plt.show() ## plt.show()를 사용하지 않으면 그래프는 똑같이 출력되나 plt.axis()의 값인 [0, 6, 0, 20]이 출력되므로 깔끔한 plot을 원한다면 plt.show()를 하자
## plt.axis()없이 plt.plot()만 해도 위의 그래프와 같이 type이 출력되므로 plt.show()를 하자
```

## 여러 개의 그래프 그리기 
`-` `matplotlib`는 대개 `numpy array`를 이용

`-` `plt.plot()`에서 `(x, y, 스타일)`을 `여러쌍 입력`하면 `한 화면`에 `여러 개의 그래프`가 그려짐

```python
import numpy as np
x = np.arange(0, 5, 0.2) ## 0부터 5까지 0.2간격으로 객체 생성

# 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(x, x, 'r--')
plt.plot(x, x**2, 'bs')
plt.plot(x, x**3, 'g^')

## plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^') 도 동일한 결과를 나타냄

plt.show()
```

`-` `plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^')` 도 동일한 결과를 나타냄

# 05. matplotlib 그래프 영역 채우기 
`-` 그래프의 `특정 영역`을 `색상`으로 `채워서` 강조 가능

`-` 3가지 함수가 존재: `fill_between()`, `fill_betweenx()`, `fill()`

`-` [$\star$] `x, y의 인덱스를 다르게 해도 되지만 범위가 맞지 않아 그래프와 무관한 사각형을 그린다` [$\star$]

## 기본 사용 

`-` `alpha는 투명도`, `color는 색상`

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, 'ko--')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y[1:3], color = 'red', alpha = 0.5) ## fill_between() 사용

plt.show()
```

`-` `fill_between()` 함수에 `x[1:3]`, `y[1:3]`를 순서대로 입력하면, 네 점 `(x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)`을 잇는 영역이 채워짐

`-` 일반화 해보면?

`-` `fill_between()` 함수에 `x[a:a+2]`, `y[b:b+2]`를 순서대로 입력하면, 네 점 `(x[a], y[a]), (x[a+1], y[a+1]), (x[a], 0), (x[a+1], 0)`을 잇는 영역이 채워짐

`-` x축에 대한 적분이라 생각하자

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, 'g*:')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[1:3], x[1:3], color ='cyan', alpha = 0.5)  ## fill_betweenx() 사용 

plt.show()
```

`-` `fill_betweenx()` 함수에 `x[1:3]`, `y[1:3]`를 순서대로 입력하면, 네 점 `(x[1], y[1]), (x[2], y[2]), (0, y[1]), (0, y[2])`을 잇는 영역이 채워짐

`-` 일반화 해보면?

`-` `fill_betweenx()` 함수에 `x[a:a+2]`, `y[a:a+2]`를 순서대로 입력하면, 네 점 `(x[a], y[a]), (x[a+1], y[a+1]), (0, y[a]), (0, y[a+1])`을 잇는 영역이 채워짐

`-` y축에 대한 적분이라 생각하자

## 두 그래프 사이 영역 채우기 
`-` `fill_between()`을 사용하면 `x축`에 대한 적분, `fill_betweenx()`를 사용하면 `y축`에 대한 적분과 같이 두 그래프 사이 영역을 채울 수 있음

### `fill_between()`

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 1, 8]

plt.plot(x1, y1, 'rd--')
plt.plot(x1, y2, 'bd--')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x1[1:3], y1[1:3], y2[1:3], color = 'gray', alpha = 0.5) ## x1, y1, y2순서 매우 중요

plt.show()
```

`-` `x축 기준으로` 두 개의 `그래프 커브 사이 영역`을 채우기 위해서는 `하나`의 값 `x1`과 `두 개`의 y 값 시퀀스 `y1, y2`를 입력

`-` `네 점 (x1[1], y1[1]), (x1[1], y2[1]), (x1[2], y1[2]), (x1[2], y2[2])` 사이 영역이 채워짐

`-` 일반화하면 `네 점 (x1[a], y1[a]), (x1[a], y2[a]), (x1[a+1], y1[a+1]), (x1[a+1], y2[a+1])` 사이 영역이 채워짐

`-` `x1, y1, y2순서` 매우 중요 ---> `한 개`의 값 `x1`을 `앞에 배치`

### `fill_betweenx()` 

```python
x1 = [1, 4, 9, 16]
x2 = [1, 2, 4, 8]
y1 = [1, 2, 3, 4]

plt.plot(x1, y, 'rd--')
plt.plot(x2, y, 'bd--')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[1:3], x1[1:3], x2[1:3], color = 'gray', alpha = 0.5) ## x1, x2, y1순서 매우 중요

plt.show()
```

`-` `y축 기준으로` 두 개의 `그래프 커브 사이 영역`을 채우기 위해서는 `하나`의 값 `y1`와 `두 개`의 x 값 시퀀스 `x1, x2`를 입력

`-` `네 점 (x1[1], y1[1]), (x2[1], y1[1]), (x1[2], y1[2]), (x2[2], y1[2])` 사이 영역이 채워짐

`-` 일반화하면 `네 점 (x1[a], y1[a]), (x2[a], y1[a]), (x1[a+1], y1[a+1]), (x2[a+1], y1[a+1])` 사이 영역이 채워짐

`-` `y1, x1, x2`순서 매우 중요 ---> `한 개`의 값 `y1`을 `앞에 배치`

## 임의의 영역 채우기 
`-` `fill()` 함수에 `x, y 값의 리스트`를 입력해주면, `각 x, y 점들로 정의되는 다각형 영역`을 자유롭게 지정해서 `채운다`

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x1, y1, 'go--')
plt.plot(x1, y2, 'bo--')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill([1.9, 1.9, 3.1, 3.1], [2, 5, 11, 8], color = 'lightgray', alpha = 0.5)

plt.show()
```

# 06. matplotlib 그리드 설정
`-` 데이터의 위치를 더 명확하게 하기 위해 그래프에 `grid()` 함수를 이용해서 `그리드(격자)를 표시`할 수 있음

`-` `plt.grid(True, axis = 'type)`와 같이 설정하면, 그래프의 x, y축에 대해 그리드가 표시

`-` `type`은 `'x'`, `'y'`, `'both'(default)` 존재

`-` `grid()`도 `선의 스타일` 설정 가능

`-` `which 파라미터`를 `'major'`, `'minor'`, `'both'` 등으로 사용하면 `주눈금`, `보조눈금`에 `각각 그리드`를 `표시`할 수 있음 ---> 나중에 더 자세히

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo--')
plt.plot(x, x**2, 'ro--', linewidth = 2)
plt.plot(x, x**3, 'go--', markersize = 5)
plt.grid(True, axis = 'both', color = 'k', alpha = 0.5, linestyle = '--', which = 'major')

plt.show()
```

# 07. matplotlib 눈금 표시
`-` `tick`은 `그래프의 축`에 `간격`을 `구분`하기 위해 표시하는 `눈금`

## 기본 사용
`-` `xticks()`, `yticks()` 함수는 각각 `x축`, `y축`에 `눈금`을 표시

`-`  `리스트` 또는 `numpy array`를 입력하면, `해당하는 위치`에 `눈금과 숫자 레이블`이 표시

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'b^')
plt.plot(a, a**2, 'r^', linewidth = 2)
plt.plot(a, a**3, 'g^', markersize = 7)

plt.xticks([0, 1, 2])
plt.yticks(np.arange(1, 6))

plt.show()
```

## 눈금 레이블 지정 
`-` `labels 파라미터`를 사용하면 `눈금 레이블`을 `명시적으로 지정`

```python
x = np.arange(0, 2.4, 0.2)

plt.plot(x, x, 'b*--')
plt.plot(x, x**2, 'r*--', linewidth = 2)
plt.plot(x, x**3, 'g*--', markersize = 7)
plt.xticks(x, labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul',  'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.yticks(np.arange(0,12), ('1M', '2M', '3M', '4M', '5M', '6M', '7M', '8M', '9M', '10M', '11M', '12M')) ## labels 없어도 적용됨

plt.show()
```

## 눈금 스타일 설정 
- `tick_params()` 함수를 사용하면 `눈금의 스타일`을 다양하게 `설정` ---> `R`의 `ggplot2`의 `axe()`와 비슷

`-` `axis`는 설정이 적용될 `축`을 지정 ---> `'x'`, `'y'`, `'both'` 중 선택

`-` `direction`은 눈금을 안/밖으로 표시 ---> `'in'`, `'out'`, `'inout'` 중 선택

`-` `length`는 `눈금의 길이`를 지정

`-` `pad`는 `눈금과 레이블과의 거리`를 지정

`-` `labelsize`는 `레이블의 크기`를 지정

`-` `labelcolor`는 `레이블의 색상`을 지정

`-` `top/bottom/left/right`를 `True/False`로 지정하면 `눈금이 표시될 위치`를 선택 ---> `default: left, bottom`

`-` `width`는 `눈금의 너비`를 지정

`-` `color`는 `눈금의 색상`을 지정

```python
x = np.arange(0, 2.4, 0.2)

plt.plot(x, x, 'b*--')
plt.plot(x, x**2, color = '#e35f62', marker = '*', linestyle = '--', linewidth=2)
plt.plot(x, x**3, color = 'springgreen', marker = '*', linestyle = '--', markersize = 7)
plt.xticks(x, labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul',  'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.yticks(np.arange(0,12), ('1M', '2M', '3M', '4M', '5M', '6M', '7M', '8M', '9M', '10M', '11M', '12M')) ## labels 없어도 적용됨

plt.tick_params(axis = 'y', direction = 'in', length = 3, pad = 6, labelsize = 11, labelcolor = 'red',  right = True)
plt.tick_params(axis = 'x', direction = 'inout', length = 10, pad = 15, labelsize = 12, width = 2, color = 'b', top = True)

plt.show()
```

# 08. matplotlib 타이틀 설정 
`-` `title()` 함수를 이용해서 `그래프의 제목`을 설정

## 기본 사용 

`-` `loc`를 통해 `상단` 중 `제목의 위치`를 `지정` 가능 ---> `left`, `right`, `center(default)` 

`-` `pad`는 `title`과 `그래프와의 간격`(오프셋)을 `포인트 단위`로 설정

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro--')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('$x^2$ plot', loc = 'right', pad = 10)  ## pad는 title과 그래프와의 간격을 포인트 단위로 설정

plt.show()
```

# 09. matplotlib 수직선/수평선 표시
`-` `그래프`의 특정 값에 해당하는 위치에 `수직선/수평선`을 표시하기 위해서 `axhline()`, `axvline()`, `hlines()`, `vlines()` 사용

## 기본 사용 - `axhline()/axvline()`
`-` `axhline()` 함수의 1번째 인자는 `y 값` 으로서 `수평선의 위치`

`-` 2, 3번째 인자는 `xmin`, `xmax 값` 으로서 `0에서 1 사이`의 값

`-` `0은 왼쪽 끝(y축)`, `1은 오른쪽 끝`을 의미

`-` `axvline()` 함수의 1번째 인자는 `x 값` 으로서 `수직선의 위치`

`-` 2, 3번째 인자는 `ymin`, `ymax 값` 으로서 `0에서 1 사이`의 값

`-` `0은 아래쪽 끝(x축)`, `1은 위쪽 끝`을 의미

`-` 2, 3번째 인자를 생략하면 `직선`을 그려줌(default) + 0\~1사이의 값은 `비율`임 ---> ex) `0.5는 중간`

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro--')

plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('$x^2$ plot', loc = 'center', pad = 15)  

plt.axhline(4, 0, 0.35, color = 'gray', linestyle = '--', linewidth = 2)
plt.axvline(2, 0, 0.23, color = 'gray', linestyle = '--', linewidth = 2)

plt.show()
```

## 기본 사용 - `hlines()/vlines()` 
`-` `axhline()/axvline()`는 `수직선/수평선`을 `비율`로 그리는 반면 `hlines()/vlines()` 는 `좌표`를 지정해서 그림

`-` `hlines()` 함수에 `y`, `xmin`, `xmax`를 순서대로 입력하면 `점 (xmin, y)`에서 `점 (xmax, y)`를 따라 `수평선`을 표시

`-` `vlines()` 함수에 `x`, `ymin`, `ymax`를 순서대로 입력하면 `점 (x, ymin)`에서 `점 (x, ymax)`를 따라 `수직선`을 표시

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'bo--')

plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('$x^2$ plot', loc = 'center', pad = 15)  

plt.hlines(9, 2, 3, color = 'k', linestyle = '--', linewidth = 2)
plt.vlines(2, 4, 9, color = 'k', linestyle = '--', linewidth = 2)

plt.show()
```

# 10. matplotlib 막대 그래프 
`-`  `bar()` 함수를 이용해서 막대 그래프를 표현

## 기본 사용 
`-` `np.arange()`는 주어진 `범위`와 `간격`에 따라 균일한 값을 갖는 `배열`을 생성하는 함수

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2019', '2020', '2021']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
```

## 스타일 꾸미기 
`-` `width`는 `막대의 너비`(default는 0.8)

`-` `align`은 `tick`과 `막대의 위치`를 `조절`

`-` `color`는 `막대의 색`

`-` `edgecolor`는 `막대의 테두리 색`

`-` `linewidth`는 `테두리의 두께`

`-` `tick_label`을 array 형태로 지정하면 `tick`에 array의 `문자열`을 나타냄

`-` `log = True`로 설정하면, `y축`이 `로그 스케일`로 표시

```python
x = np.arange(3)
years = ['2019', '2020', '2021']
values = [100, 400, 900]

plt.bar(x, values, width = 0.6, align = 'edge', color = 'k', edgecolor = 'b', linewidth = 1.5, tick_label = years, log = True)
plt.show()
```

# 11. matplotlib 수평 막대 그래프
`-` `barh()` 함수를 이용하면 `수평 막대 그래프`를 그림

## 기본 사용 
`-` `height`는 `막대의 높이`(default는 0.8)

`-` 나머지는 `bar()` 함수와 동일

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
years = ['2019', '2020', '2021']
values = [100, 400, 900]

plt.barh(x, values, height = 0.6, align = 'edge', color = 'k', edgecolor = 'b', linewidth = 1.5, tick_label = years, log = False)
plt.show()
```

# 12. matplotlib 산점도
`-` `산점도 (scatter plot)`는 두 변수의 `상관 관계`를 직교 좌표계의 평면에 데이터를 `점`으로 표현하는 그래프

`-` `scatter()` 함수를 이용하여 `산점도`를 그림

`-` `np.random.rand(N)` --->  `0~1사이의 난수 N개 생성`

`-` `scatter()`에 `x`, `y` 위치를 `입력`

`-` `s`는 `마커의 면적`을  `c`는 `마커의 색`을 지정

```python
np.random.seed(2021)  # Reproducible random state

N = 50
x = np.random.rand(N)  
y = np.random.rand(N)  

colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s = area, c = colors, alpha = 0.5)
plt.show()
```
