# matplotlib 기본 사용
`-` `matplotlib` 라이브러리를 이용해서 그래프를 그려보자

`-` 그 중에서도 `pyplot` 사용할 것 임

## 기본 그래프 

`-` `pyplot`을 이용해서 간단하게 시각화 가능

`-` `plt.plot()` 에 `하나의 숫자 리스트`를 입력함으로써 아래와 같은 그래프가 그려짐

`-` `plt.plot()`은 `하나의 리스트`만 `입력`되면 `리스트의 값들이 y 값`들이라고 가정하고, `x 값 [0, 1, 2, 3] (default)`을 `자동으로 만듦`

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.show()
```

`-` `plot()` 함수는 다양한 기능을 포함하고 있어서, `임의의 개수`의 `인자`를 `받을 수 있음`

`-` 아래와 같이 `두 개의 리스트`를 입력하면 `순서대로 x, y값`으로 생각하고 그래프를 나타냄

```python
plt.plot([1, 2, 3, 4], [10, 20, 30, 40])
```

`-` `plt.show()` 는 `화면에 표시`하는 기능을 하는데 `Jupyter`를 사용할 때는 `자동으로 표시`되기 때문에 호출할 필요가 없음

## 스타일 지정하기 
`-` `x, y 값 인자`에 대해 `선의 색상`과 `형태`를 지정하는 `포맷 문자열`을 `세번째 인자`에 `입력`할 수 있음 ---> 후에 더 자세히 알아보자

`-` `axis()`를 이용해서 축의 범위 `[xmin, xmax, ymin, ymax]`를 지정함

`-` `linewidth`를 통해 `선 두께` 조절

`-` `markersize`를 통해 `마커 사이즈` 조절

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro--', linewidth = 3, markersize = 9)
plt.axis([0, 6, 0, 20])
plt.show() ## plt.show()를 사용하지 않으면 그래프는 똑같이 출력되나 plt.axis()의 값인 [0, 6, 0, 20]이 출력되므로 깔끔한 plot을 원한다면 plt.show()를 하자
## plt.axis()없이 plt.plot()만 해도 위의 그래프와 같이 type이 출력되므로 plt.show()를 하자
```

## 여러 개의 그래프 그리기 
`-` `matplotlib`는 대개 `numpy array`를 이용

`-` `plt.plot()`에서 `(x, y, 스타일)`을 `여러쌍 입력`하면 `한 화면`에 `여러 개의 그래프`가 그려짐

```python
import numpy as np
x = np.arange(0, 5, 0.2) ## 0부터 5까지 0.2간격으로 객체 생성

# 빨간 대쉬, 파란 사각형, 녹색 삼각형
plt.plot(x, x, 'r--')
plt.plot(x, x**2, 'bs')
plt.plot(x, x**3, 'g^')

## plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^') 도 동일한 결과를 나타냄

plt.show()
```

`-` `plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^')` 도 동일한 결과를 나타냄

# matplotlib 범례 표시
`-` `범례(legend)`는 `그래프`에 `데이터의 종류`를 `표시`하기 위한 `텍스트`

## 기본 사용

`-` plot() 함수에 `label 문자열`을 지정하고 `plt.legend()` 함수를 호출하여 `범례` 표시

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label = 'Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()

plt.show()
```

## 범례 위치 지정 

`-` legend()함수의 `loc` 파라미터를 이용해서 `범례`가 `표시될 위치`를 설정 ---> 직사각형(범례 상자)의 왼쪽 아래점의 위치

`-` loc 파라미터를 `숫자 쌍 튜플`로 지정하면 해당하는 위치에 범례가 표시됨 ---> ex: (0, 0)은 좌하귀 (1, 1)은 우상귀

`-` loc 파라미터는 문자열로도 지정 가능 ---> loc = 'lower right'로 지정하면 오른쪽 아래에 범례가 표시

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label = 'price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc = (0.5, 0.5))

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label='Price ($)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc = 'lower right')

plt.show()
```

## 열 개수 지정 

`-` `ncol` 파라미터는 `범례`에 표시될 텍스트의 `열의 개수`를 지정 ---> 기본 값은 1

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label = 'Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label = 'Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc = 'best', ncol = 2) 

plt.show()
```

## 범례 테두리 꾸미기 

`-` `fontsize` 파라미터는 범례에 표시될 `폰트의 크기`를 지정

`-` `frameon` 파라미터는 범례 텍스트 상자의 `테두리 표시` 여부 지정

`-` `shadow` 파라미터는 텍스트 상자에 그림자를 표시할지 말지를 지정 

`-` 참고: [plt.legend](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [2, 3, 5, 10], label = 'Price ($)')
plt.plot([1, 2, 3, 4], [3, 5, 9, 7], label = 'Demand (#)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend(loc = 'best', ncol = 1, fontsize = 12, frameon = True, shadow = True)

plt.show()
```

# matplotlib 그래프 영역 채우기 
`-` 그래프의 `특정 영역`을 `색상`으로 `채워서` 강조 가능

`-` 3가지 함수가 존재: `fill_between()`, `fill_betweenx()`, `fill()`

`-` [$\star$] `x, y의 인덱스를 다르게 해도 되지만 범위가 맞지 않아 그래프와 무관한 사각형을 그린다` [$\star$]

## 기본 사용 

`-` `alpha는 투명도`, `color는 색상`

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, 'ko--')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y[1:3], color = 'red', alpha = 0.5) ## fill_between() 사용

plt.show()
```

`-` `fill_between()` 함수에 `x[1:3]`, `y[1:3]`를 순서대로 입력하면, 네 점 `(x[1], y[1]), (x[2], y[2]), (x[1], 0), (x[2], 0)`을 잇는 영역이 채워짐

`-` 일반화 해보면?

`-` `fill_between()` 함수에 `x[a:a+2]`, `y[b:b+2]`를 순서대로 입력하면, 네 점 `(x[a], y[a]), (x[a+1], y[a+1]), (x[a], 0), (x[a+1], 0)`을 잇는 영역이 채워짐

`-` x축에 대한 적분이라 생각하자

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, 'g*:')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[1:3], x[1:3], color ='cyan', alpha = 0.5)  ## fill_betweenx() 사용 

plt.show()
```

`-` `fill_betweenx()` 함수에 `x[1:3]`, `y[1:3]`를 순서대로 입력하면, 네 점 `(x[1], y[1]), (x[2], y[2]), (0, y[1]), (0, y[2])`을 잇는 영역이 채워짐

`-` 일반화 해보면?

`-` `fill_betweenx()` 함수에 `x[a:a+2]`, `y[a:a+2]`를 순서대로 입력하면, 네 점 `(x[a], y[a]), (x[a+1], y[a+1]), (0, y[a]), (0, y[a+1])`을 잇는 영역이 채워짐

`-` y축에 대한 적분이라 생각하자

## 두 그래프 사이 영역 채우기 
`-` `fill_between()`을 사용하면 `x축`에 대한 적분, `fill_betweenx()`를 사용하면 `y축`에 대한 적분과 같이 두 그래프 사이 영역을 채울 수 있음

### `fill_between()`

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 1, 8]

plt.plot(x1, y1, 'rd--')
plt.plot(x1, y2, 'bd--')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x1[1:3], y1[1:3], y2[1:3], color = 'gray', alpha = 0.5) ## x1, y1, y2순서 매우 중요

plt.show()
```

`-` `x축 기준으로` 두 개의 `그래프 커브 사이 영역`을 채우기 위해서는 `하나`의 값 `x1`과 `두 개`의 y 값 시퀀스 `y1, y2`를 입력

`-` `네 점 (x1[1], y1[1]), (x1[1], y2[1]), (x1[2], y1[2]), (x1[2], y2[2])` 사이 영역이 채워짐

`-` 일반화하면 `네 점 (x1[a], y1[a]), (x1[a], y2[a]), (x1[a+1], y1[a+1]), (x1[a+1], y2[a+1])` 사이 영역이 채워짐

`-` `x1, y1, y2순서` 매우 중요 ---> `한 개`의 값 `x1`을 `앞에 배치`

### `fill_betweenx()` 

```python
x1 = [1, 4, 9, 16]
x2 = [1, 2, 4, 8]
y1 = [1, 2, 3, 4]

plt.plot(x1, y, 'rd--')
plt.plot(x2, y, 'bd--')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[1:3], x1[1:3], x2[1:3], color = 'gray', alpha = 0.5) ## x1, x2, y1순서 매우 중요

plt.show()
```

`-` `y축 기준으로` 두 개의 `그래프 커브 사이 영역`을 채우기 위해서는 `하나`의 값 `y1`와 `두 개`의 x 값 시퀀스 `x1, x2`를 입력

`-` `네 점 (x1[1], y1[1]), (x2[1], y1[1]), (x1[2], y1[2]), (x2[2], y1[2])` 사이 영역이 채워짐

`-` 일반화하면 `네 점 (x1[a], y1[a]), (x2[a], y1[a]), (x1[a+1], y1[a+1]), (x2[a+1], y1[a+1])` 사이 영역이 채워짐

`-` `y1, x1, x2`순서 매우 중요 ---> `한 개`의 값 `y1`을 `앞에 배치`

## 임의의 영역 채우기 
`-` `fill()` 함수에 `x, y 값의 리스트`를 입력해주면, `각 x, y 점들로 정의되는 다각형 영역`을 자유롭게 지정해서 `채운다`

```python
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x1, y1, 'go--')
plt.plot(x1, y2, 'bo--')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill([1.9, 1.9, 3.1, 3.1], [2, 5, 11, 8], color = 'lightgray', alpha = 0.5)

plt.show()
```

# matplotlib 그리드 설정
`-` 데이터의 위치를 더 명확하게 하기 위해 그래프에 `grid()` 함수를 이용해서 `그리드(격자)를 표시`할 수 있음

`-` `plt.grid(True, axis = 'type)`와 같이 설정하면, 그래프의 x, y축에 대해 그리드가 표시

`-` `type`은 `'x'`, `'y'`, `'both'(default)` 존재

`-` `grid()`도 `선의 스타일` 설정 가능

`-` `which 파라미터`를 `'major'`, `'minor'`, `'both'` 등으로 사용하면 `주눈금`, `보조눈금`에 `각각 그리드`를 `표시`할 수 있음 ---> 나중에 더 자세히

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo--')
plt.plot(x, x**2, 'ro--', linewidth = 2)
plt.plot(x, x**3, 'go--', markersize = 5)
plt.grid(True, axis = 'both', color = 'k', alpha = 0.5, linestyle = '--', which = 'major')

plt.show()
```

# matplotlib 눈금 표시
`-` `tick`은 `그래프의 축`에 `간격`을 `구분`하기 위해 표시하는 `눈금`

## 기본 사용
`-` `xticks()`, `yticks()` 함수는 각각 `x축`, `y축`에 `눈금`을 표시

`-`  `리스트` 또는 `numpy array`를 입력하면, `해당하는 위치`에 `눈금과 숫자 레이블`이 표시

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'b^')
plt.plot(a, a**2, 'r^', linewidth = 2)
plt.plot(a, a**3, 'g^', markersize = 7)

plt.xticks([0, 1, 2])
plt.yticks(np.arange(1, 6))

plt.show()
```

## 눈금 레이블 지정 
`-` `labels 파라미터`를 사용하면 `눈금 레이블`을 `명시적으로 지정`

```python
x = np.arange(0, 2.4, 0.2)

plt.plot(x, x, 'b*--')
plt.plot(x, x**2, 'r*--', linewidth = 2)
plt.plot(x, x**3, 'g*--', markersize = 7)
plt.xticks(x, labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul',  'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.yticks(np.arange(0,12), ('1M', '2M', '3M', '4M', '5M', '6M', '7M', '8M', '9M', '10M', '11M', '12M')) ## labels 없어도 적용됨

plt.show()
```

## 눈금 스타일 설정 
- `tick_params()` 함수를 사용하면 `눈금의 스타일`을 다양하게 `설정` ---> `R`의 `ggplot2`의 `axe()`와 비슷

`-` `axis`는 설정이 적용될 `축`을 지정 ---> `'x'`, `'y'`, `'both'` 중 선택

# matplotlib 수직선/수평선 표시
`-` `그래프`의 특정 값에 해당하는 위치에 `수직선/수평선`을 표시하기 위해서 `axhline()`, `axvline()`, `hlines()`, `vlines()` 사용

## 기본 사용 - `axhline()/axvline()`
`-` `axhline()` 함수의 1번째 인자는 `y 값` 으로서 `수평선의 위치`

`-` 2, 3번째 인자는 `xmin`, `xmax 값` 으로서 `0에서 1 사이`의 값

`-` `0은 왼쪽 끝(y축)`, `1은 오른쪽 끝`을 의미

`-` `axvline()` 함수의 1번째 인자는 `x 값` 으로서 `수직선의 위치`

`-` 2, 3번째 인자는 `ymin`, `ymax 값` 으로서 `0에서 1 사이`의 값

`-` `0은 아래쪽 끝(x축)`, `1은 위쪽 끝`을 의미

`-` 2, 3번째 인자를 생략하면 `직선`을 그려줌(default) + 0\~1사이의 값은 `비율`임 ---> ex) `0.5는 중간`

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro--')

plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('$x^2$ plot', loc = 'center', pad = 15)  

plt.axhline(4, 0, 0.35, color = 'gray', linestyle = '--', linewidth = 2)
plt.axvline(2, 0, 0.23, color = 'gray', linestyle = '--', linewidth = 2)

plt.show()
```

## 기본 사용 - `hlines()/vlines()` 
`-` `axhline()/axvline()`는 `수직선/수평선`을 `비율`로 그리는 반면 `hlines()/vlines()` 는 `좌표`를 지정해서 그림

`-` `hlines()` 함수에 `y`, `xmin`, `xmax`를 순서대로 입력하면 `점 (xmin, y)`에서 `점 (xmax, y)`를 따라 `수평선`을 표시

`-` `vlines()` 함수에 `x`, `ymin`, `ymax`를 순서대로 입력하면 `점 (x, ymin)`에서 `점 (x, ymax)`를 따라 `수직선`을 표시

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'bo--')

plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('$x^2$ plot', loc = 'center', pad = 15)  

plt.hlines(9, 2, 3, color = 'k', linestyle = '--', linewidth = 2)
plt.vlines(2, 4, 9, color = 'k', linestyle = '--', linewidth = 2)

plt.show()
```

- 3차원 산점도

`-` `3D axes`를 만들기 위해 `add_subplot()`에 `projection = '3d'`를 입력

`-` `scatter()` 함수에 `x, y, z 위치`를 `array` 형태로 입력

`-` `cmap = 'Greens'`를 통해 `colormap`을 `녹색 계열`로 설정

```python
import matplotlib.pyplot as plt
import numpy as np

n = 100
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 20, 0, 50
cmin, cmax = 0, 2
x = np.array([(xmax - xmin) * np.random.random_sample() + xmin for __ in range(n)])
y = np.array([(ymax - ymin) * np.random.random_sample() + ymin for __ in range(n)])
z = np.array([(zmax - zmin) * np.random.random_sample() + zmin for __ in range(n)])
color = (cmax - cmin) * np.random.rand(n) + cmin

plt.rcParams["figure.figsize"] = (6, 6)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x, y, z, c = color, marker = 'o', s = 15, cmap = 'Greens')
plt.show()
```

## 기본 사용 
`-` 각 영역의 `비율`과 `이름`을 `ratio`와 `labels`로 지정

`-` `autopct`는 부채꼴 안에 표시될 `숫자의 형식`을 지정

`-` `autopct`에서 `문자열`로 사용할 땐 `'%' 앞`에 `백분율을 표시`하고 `'%' 뒤`에 백분율을 표시할 `형식`을 지정

`-` `'%'` 문자 자체를 출력하고 싶다면 `'%%'`와 같이 입력

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

fig = plt.figure() ## 캔버스 생성
fig.set_facecolor('white') ## 캔버스 색상 하얀색

plt.pie(ratio, labels = labels, autopct = '%.1f%%') ## autopct 문자열로 입력
plt.show()
```

## 매개 변수 사용
`-` `explode`는 `부채꼴`이 `파이 차트의 중심`에서 `벗어나는 정도`를 설정

`-` `startangle`는 `부채꼴`이 그려지는 `시작 각도`를 설정(디폴트는 0도) 

`-` `counterclock = False`로 설정하면 `시계 방향 순서`로 `부채꼴 영역`이 `표시`

`-` `shadow = True`로 설정하면 `파이 차트`에 `그림자가 표시` 

`-` `colors`를 사용하면 `각 영역의 색상`을 지정 ---> `Hex code`를 이용해서 `더욱 다양한 색상`을 지정

`-` `wedgeprops`는 `부채꼴 영역`의 `스타일`을 `설정`

```python
import matplotlib.pyplot as plt

ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]
colors = ['#ff9999', '#ffc000', 'green', 'lightgray']
wedgeprops = {'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}

fig = plt.figure()
fig.set_facecolor('white')

plt.pie(ratio, labels = labels, autopct = lambda x: '{:.1f}%'.format(x), startangle = 260, counterclock = False, explode = explode, colors = colors, wedgeprops = wedgeprops, shadow = True) ## autopct 함수로 입력
plt.show()
```

# matplotlib Errorbar
`-` `errorbar()` 함수를 이용해서 그래프에 `에러바(errorbar, 오차막대)`를 나타냄

## 기본 사용 
`-` `yerr`의 각 값들은 데이터 포인트의 `위/아래 대칭`인 `오차`로 표시됨

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [2.3, 3.1, 1.7, 2.5] ## 표준편차

plt.errorbar(x, y, yerr = yerr) ## yerr ---> y_error 
plt.show()
```

## 비대칭 편차 나타내기

`-` 데이터 포인트를 기준으로 `비대칭인 편차`를 표시하려면 `(2, N) 형태의 값`들을 입력

`-` `첫번째 튜플`의 값들은 `아래 방향 편차`이고 `두번째 튜플`의 값들은 `위 방향 편차`임

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
yerr = [(2.3, 3.1, 1.7, 2.5), (1.1, 2.5, 0.9, 3.9)] ## 첫번째 튜플은 아래 방향편차, 두번째 튜플은 위 방향 편차

plt.errorbar(x, y, yerr = yerr, color = 'red')
plt.show()
```

## 상한/하한 기호 표시

`-` `uplims`와 `lolims`를 사용해서 `상한/하한 기호`를 선택적으로 `표시`

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 5)
y = x**2
yerr = np.linspace(0.1, 0.4, 4)

plt.errorbar(x, y + 4, yerr = yerr)
plt.errorbar(x, y + 2, yerr = yerr, uplims = True, lolims = True)

upperlimits = [True, False, True, False]
lowerlimits = [False, False, True, True]
plt.errorbar(x, y, yerr = yerr, uplims = upperlimits, lolims = lowerlimits)
plt.show()
```

`-` `plt.text()` 함수는 `그래프`의 적절한 위치에 `텍스트`를 `삽입`함

## 기본 사용 

`-` `text()`에 그래프 상의 `x 위치`와 `y 위치`와 `삽입할 텍스트`를 순서대로 입력

```python
import matplotlib.pyplot as plt
import numpy as np

a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins = 100, density = True, alpha = 0.7, histtype = 'step', color = 'red')
plt.hist(b, bins = 50, density = True, alpha = 0.5, histtype = 'stepfilled', color = 'blue')
plt.hist(c, bins = 100, density = True, alpha = 0.9, histtype = 'step', color = 'green')

plt.text(1.0, 0.35, '2.0*np.random.randn(10000) + 1.0')
plt.text(2.0, 0.20, 'np.random.standard_normal(10000)')
plt.text(5.0, 0.08, 'np.random.rand(5000) - 10.0')

plt.show()
```

## 텍스트\텍스트 상자 스타일 설정 

`-` `fontdict` 키워드를 이용하면 `font`의 `종류`, `크기`, `색상`, `투명도`, `weight` 등의 `텍스트 스타일`을 설정 가능

`-` `rotation` 파라미터를 이용해서 `텍스트`를 `회전`

`-` `bbox`를 사용해서 `텍스트 상자`의 `스타일`을 설정

`-` 참고: [텍스트 속성](https://matplotlib.org/3.3.1/tutorials/text/text_props.html)

|Property|Description|Option|
|:-:|:-:|:-:|
|alpha|텍스트의 투명도|0.0 ~ 1.0 (float)|
|color|텍스트의 색상|Any Matplotlib color|
|family|텍스트의 글꼴|FONTNAME: 'serif', 'sans-serif', 'cursive', 'fantasy', 'monospace'|
|rotation|텍스트의 회전각|angle in degrees, 'vertical', 'horizontal'|
|size|텍스트의 크기|size in points, etc|
|weight|텍스트의 굵기|0-1000(numeric), etc|

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

font1 = {'family': 'serif',
      'color':  'blue',
      'weight': 'normal',
      'size': 15}

font2 = {'family': 'Times New Roman',
      'color':  'darkred',
      'weight': 'bold',
      'size': 12,
      'alpha': 0.7}

font3 = {'family': 'Arial',
      'color':  'forestgreen',
      'style': 'italic',
      'size': 14}

box1 = {'boxstyle': 'square',     ## 'ec'는 'edgecolor', 'fc'는 'facecolor'
        'ec': (0.5, 0.5, 1.0),
        'fc': (0.8, 0.8, 1.0),
        'linestyle': '--'}

box2 = {'boxstyle': 'round',
        'ec': (1.0, 0.5, 0.5),
        'fc': (1.0, 0.8, 0.8)}

box3 = {'boxstyle': 'square',
        'ec': (0.3, 1.0, 0.5),
        'fc': (0.8, 1.0, 0.5),
        'linestyle': '-.',
        'linewidth': 2}

plt.hist(a, bins = 100, density = True, alpha = 0.7, histtype = 'step', color = 'red')
plt.hist(b, bins = 50, density = True, alpha = 0.5, histtype = 'stepfilled', color = 'blue')
plt.hist(c, bins = 100, density = True, alpha = 0.9, histtype = 'step', color = 'green')

plt.text(-3.0, 0.15, 'np.random.randn()', fontdict = font1, rotation = 85, bbox = box1)
plt.text(2.0, 0.0, 'np.random.standard_normal()', fontdict = font2, rotation = -60, bbox = box2)
plt.text(-10.0, 0.08, 'np.random.rand()', fontdict = font3, bbox = box3)

plt.show()
```
